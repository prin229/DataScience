```python
df=pd.read_csv('train.csv')
df.head(3)
```

```python
pd.set_option('display.max_rows',81)
tmp=df.isnull().sum()
for i in df.columns:
    if (tmp[i]/len(df))*100 <=5:
       df.dropna(axis=0, subset=[i],inplace=True)
```

```python
df.index
```

```python
df.isnull().sum()
```

```python
df['LotFrontage'].fillna(df.LotFrontage.mean(),inplace=True)
```

```python
df['Alley'].fillna('None',inplace=True)
```

```python
tmp_dict={'Pave':3,'Grvl':2,'None':1}
df['Alley']=df['Alley'].map(tmp_dict)
```

```python
df['FireplaceQu'].fillna('None',inplace=True)
```

```python
tmp_dict1={'None':1, 'TA':2, 'Gd':3, 'Fa':4, 'Ex':5, 'Po':6}
df['FireplaceQu']=df['FireplaceQu'].map(tmp_dict1)
```

```python
tmp=df.isnull().sum()
```

```python
df['GarageType'].unique()
```

```python
for i in df.columns:
    if df[i].dtype.name=='object':
        df[i].fillna('None',inplace=True)
```

```python
df['GarageYrBlt'].fillna(int(df.GarageYrBlt.mode()),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
df['SalePrice'].ravel()
```

```python

df['MSZoning'].unique()
```

```python
encoder=LabelEncoder()
for i in df.columns:
    if df[i].dtype.name=='object':
        tmp=encoder.fit_transform(df[i])
        df.drop(i,axis=1,inplace=True)
        df[i]=tmp
        
        
        
```

```python
from sklearn.model_selection import train_test_split
```

```python
features=[i for i in df.columns if i not in ['Id','SalePrice']]
X_train,X_test,y_train,y_test=train_test_split(df[features],df['SalePrice'],test_size=0.2,random_state=40)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_jobs=-1,n_estimators=100,max_depth=5)
model.fit(df[features],df['SalePrice'])
```

```python
test_df=pd.read_csv('test.csv')
test_df.head()
```

```python
test_df.isnull().sum()
```

```python
for i in test_df.columns:
    if test_df[i].dtype.name=='object':
        test_df[i].fillna('None',inplace=True)
        tmp=encoder.fit_transform(test_df[i])
        test_df.drop(i,axis=1,inplace=True)
        test_df[i]=tmp
    elif test_df[i].dtype.name=='int64' or test_df[i].dtype.name=='int32':
        if(((test_df[i].isnull().sum())/len(test_df))*100>5):
            test_df[i].fillna(int(test_df[i].mode()),inplace=True)
        else:
            test_df[i].fillna(test_df[i].mean(),inplace=True)
    else:
        if(((test_df[i].isnull().sum())/len(test_df))*100>5):
            test_df[i].fillna(test_df[i].mean(),inplace=True)
        else:
            test_df[i].fillna(test_df[i].mean(),inplace=True)
            
```

```python
test_df.head()
```

```python
y_predict=model.predict(test_df[features])
```

```python
tmp_dict=dict(zip(test_df.Id,y_predict))
```
